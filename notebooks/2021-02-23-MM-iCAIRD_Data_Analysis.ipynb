```python
import pandas as pd
import numpy as np
import scipy
import csv
from functools import reduce
from scipy import stats
```

```python
#Convert excel file and save it as csv file
read_file = pd.read_excel('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD Data.xls', sheet_name='iCAIRD_Data')
read_file.to_csv ('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Data.csv', index= False,  header=True)
```

```python
#split Cervical from Endometrial 
df = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Data.csv')
df['Image Filename'] = df['Image Filename'].map(lambda x: x.lstrip("export\\"))

cervical_df = df[(df['Sample']== 'Cervical')]
cervical_df = cervical_df.sort_values(by=['StudyID'])
cervical_df.index = np.arange(0, len(cervical_df))


endometrial_df = df[(df['Sample']== 'Endometrial')]
endometrial_df = endometrial_df.sort_values(by=['StudyID'])
endometrial_df.index = np.arange(0, len(endometrial_df))

#save the csv files
cervical_df.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_cervical.csv', index = False, header = True)
endometrial_df.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_endometrial.csv', index = False, header = True)


cervical_df
```

```python
#Convert txt file to csv
with open('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Samples_Details.txt', 'r', encoding='unicode_escape') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(",") for line in stripped if line)
    with open('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Samples_Details.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerows(lines)
```

```python
# read iCAIRD sample details file 
df = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Samples_Details.csv')

df1 = df[(df['SampleType']== '"CX"')]
df2 = df[(df['SampleType']== '"EN"')]

cervical_staining_info = df1[['StudyID','StainingSiteID','StainingProtocol']]
endometrial_staining_info = df2[['StudyID','StainingSiteID','StainingProtocol']]


cervical_staining_info.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/cervical_staining_info.csv',index = False, header = True)
endometrial_staining_info.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/endometrial_staining_info.csv', index = False, header = True)

cervical_staining_info
```

```python
#read staining info csv files
cervical_staining_info = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/cervical_staining_info.csv', sep = ',').replace('"','', regex=True)
endometrial_staining_info = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/endometrial_staining_info.csv', sep = ',').replace('"','', regex=True)

#save the files after removing the qutations around values
cervical_staining_info.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/cervical_staining.csv', index = False)
endometrial_staining_info .to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/endometrial_staining.csv', index = False)

#read the csv files to Dataframes
cervical_staining = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/cervical_staining.csv')
endometrial_staining = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/endometrial_staining.csv')

cervical_staining
```

```python
#join two dataframes for Cervical
result1 = pd.concat([cervical_df, cervical_staining], axis=1, join="inner")
result1.columns.values[6]= 'BiobankID'

# All samples from staining Site 6 and 8 are test samples
result1.loc[result1.StainingSiteID == 6, 'train/valid/test'] = 'test'
result1.loc[result1.StainingSiteID == 8, 'train/valid/test'] = 'test'



#join two dataframes for Endometrial
result2 = pd.concat([endometrial_df, endometrial_staining], axis=1, join="inner")
result2.columns.values[6]= 'BiobankID'

# All samples from staining Site 6 and 8 are test samples
result2.loc[result2.StainingSiteID == 6, 'train/valid/test'] = 'test'
result2.loc[result2.StainingSiteID == 8, 'train/valid/test'] = 'test'


result1
```

```python
#number of test samples from stain labs 6 and 8 for cervical and endometrial
test_cervical = len(result1[result1['train/valid/test'] == 'test'])
test_endo =  len(result2[result2['train/valid/test'] == 'test'])

print('Number of test samples form stain labs 6 & 8 is {} in Cervical data'.format(test_cervical))
print('Number of test samples form stain labs 6 & 8 is {} in Endometrial data'.format(test_endo))
```

```python
# Add 10% of samples from other staining sites to test (Balanced based of categories and stain labs)
#for Cervical
remain1 = result1[result1['train/valid/test'].isna()]
grouped1 = remain1.groupby(['Category', 'StainingSiteID']).apply(lambda x: x.sample(frac = 0.1))

grouped1
```

```python
file_names = grouped1['Image Filename'].tolist()
for name in file_names :
    result1.loc[(result1['Image Filename'] == name),'train/valid/test']='test'

result1.loc[(result1['train/valid/test'].isna()), 'train/valid/test'] = 'train'
result1
```

```python
# Add 10% of samples from other staining sites to test (Balanced based of categories and stain labs)
#for Endometrial
remain2 = result2[result2['train/valid/test'].isna()]
grouped2 = remain2.groupby(['Category', 'StainingSiteID']).apply(lambda x: x.sample(frac = 0.1))
grouped2
```

```python
file_names = grouped2['Image Filename'].tolist()
for name in file_names :
    result2.loc[(result2['Image Filename'] == name),'train/valid/test']='test'

result2.loc[(result2['train/valid/test'].isna()), 'train/valid/test'] = 'train'
result2
```

```python
train_cervical = len(result1[result1['train/valid/test'] == 'train'])
test_cervical =  len(result1[result1['train/valid/test'] == 'test'])
print('Number of train samples is {} and test samples is {} in Cervical data'.format(train_cervical, test_cervical))

train_endo = len(result2[result2['train/valid/test'] == 'train'])
test_endo =  len(result2[result2['train/valid/test'] == 'test'])
print('Number of train samples is {} and test samples is {} in Endometrial data'.format(train_endo, test_endo))
```

```python
result1['Category'] = result1['Category'].str.lower()
result1['subCategory'] = result1['subCategory'].str.lower()
result1.replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
result1['subCategory'].replace({"cin_1": "cin1", "cin_2": "cin2", "cin_3":"cin3"}, inplace=True)
result1
```

```python
result2['Category'] = result2['Category'].str.lower()
result2['subCategory'] = result2['subCategory'].str.lower()
result2.replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
result2['subCategory'].replace({"hyperplasia_with_atypia": "hyperplasia"}, inplace=True)
result2
```

```python
#save final dataframes as csv files
result1.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Cervical_Data.csv', index= False, header = True)
result2.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Endometrial_Data.csv', index= False, header = True)
```
