## Find all the paths for the dataset

```python
from repath.utils.paths import project_root

annotations_dir = project_root() / 'data' / 'icaird_blood' / 'annotations'
annotation_paths = list(annotations_dir.glob('*.txt'))
len(annotation_paths)
```

```python
slides_dir = project_root() / 'data' / 'isilon1' / 'iCAIRD'
len(list(slides_dir.glob('*.isyntax')))
```

```python
from pathlib import Path

def to_slide_path(annotation_path: Path) -> Path:
    name = str(annotation_path.name)[:-9] + '.isyntax'
    return slides_dir / name
    
slide_paths = [to_slide_path(p) for p in annotation_paths]
len(slide_paths)
```

```python
paths = list(zip(annotation_paths, slide_paths))
len(paths)
```
