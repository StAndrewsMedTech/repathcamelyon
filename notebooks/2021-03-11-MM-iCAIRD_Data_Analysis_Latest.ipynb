```python
import pandas as pd
```

```python
df = pd.read_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/SampleImageData_210309.csv', encoding='cp1252')
print('Total number of sample:', len(df))
df.head()
```

```python
cervical_df = df[(df['SampleType']== 'CX')]
cervical_df['Category'] = cervical_df['Category'].str.lower()
cervical_df['subCategory'] = cervical_df['subCategory'].str.lower()
cervical_df['Category'].replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
cervical_df['subCategory'].replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
cervical_df['subCategory'].replace({"cin_1": "cin1", "cin_2": "cin2", "cin_3":"cin3"}, inplace=True)
cervical_df
```

```python
endometrial_df = df[(df['SampleType']== 'EN')]
endometrial_df['Category'] = endometrial_df['Category'].str.lower()
endometrial_df['subCategory'] = endometrial_df['subCategory'].str.lower()
endometrial_df['Category'].replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
endometrial_df['subCategory'].replace({'/':'_', ' ':'_'}, regex=True, inplace=True)
endometrial_df['subCategory'].replace({"hyperplasia_with_atypia": "hyperplasia"}, inplace=True)
endometrial_df
```

```python
cervical_df.loc[cervical_df.StainingSiteID == 6, 'train/valid/test'] = 'test'
cervical_df.loc[cervical_df.StainingSiteID == 8, 'train/valid/test'] = 'test'
# Add 10% of samples from other staining sites to test (Balanced based of categories and stain labs)
#for Cervical
remain1 = cervical_df[cervical_df['train/valid/test'].isna()]
grouped1 = remain1.groupby(['Category', 'StainingSiteID']).apply(lambda x: x.sample(frac = 0.1))
file_names = grouped1['Filename'].tolist()
for name in file_names :
    cervical_df.loc[(cervical_df['Filename'] == name),'train/valid/test']='test'
cervical_df.loc[(cervical_df['train/valid/test'].isna()), 'train/valid/test'] = 'train'
cervical_df.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Cervical_Data.csv', index = False, header = True)
#############################################################################################################
```

```python
endometrial_df.loc[endometrial_df.StainingSiteID == 6, 'train/valid/test'] = 'test'
endometrial_df.loc[endometrial_df.StainingSiteID == 8, 'train/valid/test'] = 'test'
# Add 10% of samples from other staining sites to test (Balanced based of categories and stain labs)
#for Cervical
remain2 = endometrial_df[endometrial_df['train/valid/test'].isna()]
grouped2 = remain2.groupby(['Category', 'StainingSiteID']).apply(lambda x: x.sample(frac = 0.1))
file_names = grouped2['Filename'].tolist()
for name in file_names :
    endometrial_df.loc[(endometrial_df['Filename'] == name),'train/valid/test']='test'
endometrial_df.loc[(endometrial_df['train/valid/test'].isna()), 'train/valid/test'] = 'train'
endometrial_df.to_csv('/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Endometrial_Data.csv', index = False, header = True)
```

```python
train_cervical = len(cervical_df[cervical_df['train/valid/test'] == 'train'])
test_cervical =  len(cervical_df[cervical_df['train/valid/test'] == 'test'])
print('Number of train samples is {} and test samples is {} in Cervical data'.format(train_cervical, test_cervical))

train_endo = len(endometrial_df[endometrial_df['train/valid/test'] == 'train'])
test_endo =  len(endometrial_df[endometrial_df['train/valid/test'] == 'test'])
print('Number of train samples is {} and test samples is {} in Endometrial data'.format(train_endo, test_endo))
```
