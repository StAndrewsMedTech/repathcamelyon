```python
import pandas as pd
```

```python
path = '/home/mahnaz/iCAIRD_DATA_ANALYSIS/iCAIRD_Cervical_Data.csv'
```

```python
cervical_df = pd.read_csv(path)
```

```python
cervical_train = cervical_df[cervical_df['train/valid/test']=='train']
cervical_train.head(10)
```

```python
annotations = 'iCAIRD/annotations/' + cervical_train['Image Filename'].replace({'isyntax':'txt'}, regex=True)
```

```python
df = pd.DataFrame()
df["slide"] = 'iCAIRD/'+ cervical_train['Image Filename']
df["label"] = cervical_train['Category']
df["annotation"] = annotations
df["tags"] = cervical_train['subCategory']
df.head(10)
```

```python
cervical_test = cervical_df[cervical_df['train/valid/test']=='test']
cervical_test.head(10)
```

```python
test_annotations = 'iCAIRD/annotations/' + cervical_test['Image Filename'].replace({'isyntax':'txt'}, regex=True)
```

```python
df = pd.DataFrame()
df["slide"] = 'iCAIRD/'+ cervical_test['Image Filename']
df["label"] = cervical_test['Category']
df["annotation"] = test_annotations
df["tags"] = cervical_test['subCategory']
df.head(10)
```
